# K-mean Clustering

```python
df
```

```python
df_new = df.drop(['State'], axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4,init='k-means++')
```

```python
y_predicted = km.fit_predict(df_new)
y_predicted
```

```python
df_new = pd.DataFrame(df_new)
```

```python
df_new['cluster']=y_predicted
```

```python
# WCSS Score
wcss = km.inertia_
wcss
```

```python
## Checking Score
from sklearn.metrics import silhouette_score
score = silhouette_score(df_new, y_predicted)
score
```

```python
krng = range(1,11)
wcss = []
for k in krng:
    km = KMeans(n_clusters=k)
    km.fit_predict(df_new)
    wcss.append(km.inertia_)
```

```python
plt.xlabel('k')
plt.ylabel('wcss')
plt.plot(krng, wcss);
```

```python
# Trying k=2
km = KMeans(n_clusters=2)
```

```python
y_predicted = km.fit_predict(df_new)
y_predicted
```

```python
score = silhouette_score(df_new, y_predicted)
score
```

# PCA

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(df_new)
scaled_data = scalar.transform(df_new)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
new_data = pd.DataFrame(x_pca, columns=['component_1', 'component_2', 'component_3'])
```
